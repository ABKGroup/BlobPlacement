```python
from gen_cluster_from_placement import generate_partition_bins, partition_box, generate_bins, generate_random_bins
from matplotlib import pyplot as plt
from matplotlib import patches as patches
```

```python
die = [0, 0, 323408, 321600]
ar = 0.3
min_area = 10000000
count = 150
imbalance = 0.3
bins = generate_partition_bins(die, count, min_area, ar, imbalance)
# w = partition_box(die[2], die[3], imbalance, ar, min_area)
# print(w)
```

```python
## Draw the bins and the die
fig, ax = plt.subplots()
ax.set_xlim(0, die[2])
ax.set_ylim(0, die[3])
for b in bins:
    widht = b[2] - b[0]
    height = b[3] - b[1]
    ax.add_patch(patches.Rectangle((b[0], b[1]), widht, height, fill=False))
```
